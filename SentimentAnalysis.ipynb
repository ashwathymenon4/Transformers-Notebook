```python
from transformers import pipeline
```

```python
# create your pipeline - includes tokenization etc
classifier = pipeline("sentiment-analysis")
```

```python
# no need to convert into pytorch tensor, numpy array, tensorflow tensor etc
```

```python
# output of classifier is a dictionary
classifier("This is such a great movie")
```

```python
classifier("This is not a great movie")
```

```python
# can pass list as well
classifier(["This is a great movie","This is a terrible movie"])
```
